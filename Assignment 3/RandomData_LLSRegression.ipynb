```python
import numpy as np
from numpy.linalg import inv
import matplotlib.pyplot as plt
from scipy import stats
```

```python
Y = np.random.uniform(0,20,100)
X = Y+np.random.uniform(0,20,100)
```

## Linear least squares scipy

```python
y_pred = stats.linregress(X,Y)
y_pred
```

```python
plt.scatter(X,Y)
plt.plot(X,y_pred.intercept+y_pred.slope*X,c="orange")
```

## Random Line

```python
M = np.arange(1,3,0.3)
plt.scatter(X,Y)
for i in M:
    y_pred = X * i
    plt.plot(X,y_pred)
```

```python
X = X.reshape(X.shape[0],1)
Y = Y.reshape(Y.shape[0],1)
```

## Linear least squares

```python
M = np.matmul(inv(np.matmul(X.T,X)), np.matmul(X.T,Y))
```

```python
y_pred = np.matmul(X,M)
```

```python
plt.scatter(X,Y)
plt.plot(X,y_pred,c="orange")
```
